```python
import os
import time
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
from selenium import webdriver

product = "Anouk Sarees"

# define the epath to chrome webdriver executable
# initialise the chrome webdriver 

driver = webdriver.Chrome()

# navigate the url

driver.get("https://www.myntra.com/AnoukSarees?rawquery="+product)

# once the page is loaded u can extract the page text

myntra_text = driver.page_source

myntra_html = bs(myntra_text , "html.parser")
```

```python
review_html = bs(review_page,'html.parser')
review_html
```

```python
review = review_html.find_all("div",{"class":"detailed-reviews-userReviewsContainer"})
print(review)
```

```python
for i in review:
    user_rating = i.findAll('div',{"class":"user-review-starWrapper"})
    user_comment = i.findAll('div',{"class": "user-review-reviewTextWrapper"})
    user_name = i.findAll('div',{"class": "user-review-left"})
```

```python
for i in range(len(user_rating)):
    rating = user_rating[i].find("span",class_="user-review-starRating").text
    comment = user_comment[i].text
    name =user_name[i].find('span').text
    date = user_name[i].find_all('span')[1].text
    print(f"DATE: {date} ,\nRATING: {rating} ,\nNAME: {name},\nCOMMENT: {comment}\n")
```
